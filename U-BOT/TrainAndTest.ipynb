```python
import pickle
import random
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('punkt')
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
```

```python
lemmatizer = WordNetLemmatizer()
f=open('scraped.pickle','rb')
data=pickle.load(f)
data
```

```python
import string
words=[]
classes=[]
documents=[]
ignore_ch=[char for char in string.punctuation]
ignore_ch
```

```python
for term in data['master']:
    for pattern in term['patterns']:
        word_list=word_tokenize(pattern)
        #print(word_list)
        #words.append(word_list)
        words.extend(word_list)
        documents.append((word_list,term['tag']))
        if term['tag'] not in classes:
            classes.append(term['tag'])
```

```python
nltk.download('wordnet')
words=[lemmatizer.lemmatize(word) for word in words if word not in ignore_ch]
words
```

```python
words=sorted(set(words))
classes=sorted(set(classes))
```

```python
f1=open('words.pickle','wb')
pickle.dump(words, f1)
f1.close()
```

```python
f2=open('classes.pickle','wb')
pickle.dump(classes, f2)
f2.close()
```

```python
training_data=[]
output_empty=[0]*len(classes)
```

```python
for document in documents:
    bag=[]
    word_pat=document[0]
    word_pat=[lemmatizer.lemmatize(word.lower()) for word in word_pat]
    
    for word in words:
        bag.append(1) if word in word_pat else bag.append(0)
    output_row=list(output_empty)
    output_row[classes.index(document[1])]=1
    training_data.append([bag,output_row])
print(training_data)
```

```python
random.shuffle(training_data)
training_data=np.array(training_data)
train_X=list(training_data[:,0])
train_Y=list(training_data[:,1])
```

```python
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=10)

model=Sequential()
model.add(Dense(128,input_shape=(len(train_X[0]),),activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_Y[0]),activation='softmax'))

sgd= SGD(lr=0.01, decay=1e-6,momentum=0.9,nesterov=True)

model.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])

mod=model.fit(np.array(train_X),np.array(train_Y),epochs=150, batch_size=5,verbose=1,callbacks=es)

model.save('model.h5',mod)
print('Trained')
```
