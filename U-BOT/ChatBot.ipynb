```python
import random
import pickle
import numpy as np 
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from tensorflow.keras.models import load_model
lemmatizer=WordNetLemmatizer()
file=open('scraped.pickle','rb')
data=pickle.load(file)
file1=open('words.pickle','rb')
file2=open('classes.pickle','rb')
words=pickle.load(file1)
classes=pickle.load(file2)
model =load_model('model.h5')
```

```python
def preprocess(sen):
    sen_words=nltk.word_tokenize(sen)
    sen_words=[lemmatizer.lemmatize(word.lower()) for word in sen_words]
    return sen_words
def createBag(sen):
    sen_words=preprocess(sen)
    bag=[0]*len(words)
    for w in sen_words:
        for i,word in enumerate(words):
            if word==w:
                bag[i]=1
    return np.array(bag)
```

```python
def predict_class(sen):
    res_list=[]
    bow=createBag(sen)
    res=model.predict(np.array([bow]))[0]
    thresh=0.75
    results=[[i,r] for i,r in enumerate(res) if r>thresh]
    results.sort(key=lambda x:x[1], reverse=True)
    res_list=[]
    print(results)
    for r in results:
        res_list.append({'class': classes[r[0]],'prob':str(r[1])})
    return res_list
```

```python
def getResponse(a,b):
    res=""
    tag=a[0]['class']
    lit=b['master']
    for i in lit:
        if i['tag']==tag:
            print(i['responses'])
            res=random.choice(i['responses'])
            break
    return res
```
